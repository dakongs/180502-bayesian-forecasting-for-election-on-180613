# 2. 2014년 6월 부산시장

```python
dfp_2014 = pd.read_excel('./raw_data/pusan_2014.xlsx')
```

```python
dfp_2014 = dfp_2014.loc[2:]
```

```python
dfp_2014.columns = ['id', '조사기간', '조사기관', '조사의뢰', '방법', '표본크기', '표본오차', '응답률', '서병수', '고창권', '오거돈', '기타 및 무응답']
```

```python
dfp_2014.head()
```

```python
dfp_2014 = dfp_2014.loc[3:]
```

```python
dfp_2014.head()
```

```python
dfp_2014 = dfp_2014[['조사기간', '조사기관', '조사의뢰', '표본크기', '응답률', '서병수', '고창권', '오거돈', '기타 및 무응답']]
```

```python
dfp_2014.index = range(len(dfp_2014))
```

```python
dfp_2014.head()
```

```python
def find_end(x):
    strs = str(x).split('~')
    
    if len(strs) == 1:
        return strs[0]
    else:
        return strs[1]
```

```python
dfp_2014['조사기간'] = dfp_2014['조사기간'].apply(lambda x: find_end(x))
```

```python
dfp_2014.head()
```

```python
dfp_2014 = dfp_2014.rename(columns={'조사기간': '조사끝'})
```

```python
dfp_2014.head()
```

```python
def clean_dash(x):
    if x == '-':
        return 0 
    else:
        return x
```

```python
dfp_2014['고창권'] = dfp_2014['고창권'].apply(lambda x : clean_dash(x))
```

```python
dfp_2014.head()
```

```python
dfp_2014['기타 및 무응답'] = dfp_2014['기타 및 무응답'].apply(lambda x:float(x)) + dfp_2014['고창권'].apply(lambda x: float(x))
```

```python
dfp_2014 = dfp_2014.rename(columns={'기타 및 무응답': '기타'})
```

```python
del dfp_2014['고창권']
```

```python
dfp_2014['조사끝'] = dfp_2014['조사끝'].apply(lambda x: str(140) + str(x.replace('.', '')))
```

```python
dfp_2014.head()
```

```python
dfp_2014.to_csv('./clean_data/2014_pusan.csv')
```
